```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('us_inflation_rates.csv')

# Assuming the 'date' column is in datetime format, you can convert it if necessary
# df['date'] = pd.to_datetime(df['date'])

# Filter data for the years 2010 to 2014
start_date = '1947-01-01'
end_date = '2013-12-31'

filtered_df = df[(df['date'] >= start_date) & (df['date'] <= end_date)]

# Display the filtered DataFrame
print(filtered_df)
```

```python
print(df.info())
print(df.describe())
```

```python
# Check for missing values
print(filtered_df.isnull().sum())

# If there are missing values, you can choose to drop or impute them.
# For example, to drop rows with missing values:
df.dropna(inplace=True)
```
